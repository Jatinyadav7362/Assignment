```python
i
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression 

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error



# Extract car brand from the 'Name' column
df['Brand'] = df['Name'].apply(lambda x: x.split()[0])

# Define features and target variable
X = df[['Mileage_numerical']]
y = df['Price']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train a regression model (e.g., Random Forest) on the entire dataset
model = LinearRegression()

model.fit(X_train, y_train)

# Make predictions on the entire dataset
y_pred = model.predict(X_test)

# Calculate R-squared, MSE, and MAE for the entire dataset
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

# Create a DataFrame that includes X_test, actual price (y_test), predicted price, error, and error percentage
results_df = pd.DataFrame({
    'Mileage_numerical': X_test['Mileage_numerical'],
    'Actual_Price': y_test,
    'Predicted_Price': y_pred,
})
results_df['Error'] = abs(results_df['Actual_Price'] - results_df['Predicted_Price'])
results_df['Error_Percentage'] = (results_df['Error'] / results_df['Actual_Price']) * 100

# Display the evaluation metrics
print(f'R-squared: {r2:.2f}')
print(f'Mean Squared Error: {mse:.2f}')
print(f'Mean Absolute Error: {mae:.2f}')

# Display the results DataFrame
results_df

```
